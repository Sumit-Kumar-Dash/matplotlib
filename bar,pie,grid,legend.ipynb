## bar plot

```python
import matplotlib.pyplot as plt
```

```python
plt.bar([1,2,3,4],[2,4,6,8])
plt.axis([0,6,0,12])
plt.title("plot 1")
plt.xlabel("horizontal axis")
plt.ylabel("vertical axis")
plt.show()
```

```python
plt.barh([1,2,3,4],[2,4,6,8])
plt.axis([0,6,0,12])
plt.title("plot 1")
plt.xlabel("horizontal axis")
plt.ylabel("vertical axis")
plt.show()
```

```python
plt.bar([1,2,3,4],[2,4,6,8],edgecolor="red",color="green")
```

```python
plt.bar([1,2,3,4],[2,4,6,8],edgecolor=["red","black"],color=["green","white","yellow","blue"])
```

```python
plt.bar([1,2,3,4],[2,4,6,8],edgecolor=["red","black","orange","brown"],color=["green","white","yellow","blue"])
```

```python
plt.bar([1,2,3,4],[2,4,6,8],edgecolor=["red","black","orange","brown"],color=["green","white","yellow","blue"],linewidth=4)
```

## pie chart

```python
plt.pie([10,15,18])
```

```python
plt.title("pie")
plt.xlabel("horizontal axis")
plt.ylabel("vertical axis")
plt.pie([10,15,18],labels=["python","r","java"])
```

```python
plt.pie([10,15,18],labels=["python","r","java"],startangle=90)
```

```python
plt.pie([10,15,18],labels=["python","r","java"],startangle=0)
```

```python

plt.pie([10,15,18],labels=["python","r","java"],startangle=90)
plt.legend(loc="best")
```

```python
plt.pie([10,15,18],labels=["python","r","java"],startangle=90,shadow=True)
```

```python
plt.pie([10,15,18],labels=["python","ruby","java"],startangle=90,shadow=True,labeldistance=7)
```

```python
plt.pie([10,15,18],labels=["python","r","java"],colors=["black","red","yellow"])
```

```python
plt.pie([10,15,18],labels=["python","r","java"],colors=["black","red","yellow"],explode=[0.1,0.3,0.4])
```

## grid

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.axis([0,6,0,12])
plt.title("plot 1")
plt.grid(True)
```

```python
plt.grid(linestyle="--")
```

```python
plt.grid(linestyle="-.")
```

```python
plt.grid(linestyle="--",color="red")
```

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.grid(linestyle="--",color="red",which="major")
```

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.minorticks_on()
plt.grid(linestyle="--",color="red",which="minor")
```

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.minorticks_on()
plt.grid(linestyle="-",color="green",which="major",alpha=0.9)
plt.grid(linestyle="--",color="red",which="minor",alpha=0.5)
```

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.minorticks_on()
plt.grid(linestyle="-",color="green",which="both")
```

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.grid(linestyle="--",color="red",which="major",axis='y')
```

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.grid(linestyle="--",color="red",linewidth=2.0)
```

## save the image

```python
plt.plot([1,2,3,4],[2,4,6,8],"b-")
plt.grid(linestyle="--",color="red",linewidth=2.0)
plt.savefig("new.jpg")
```

## legend

```python
x1=[1,2,3,4]
x2=[2,4,6,8]
y1=[3,6,9,12]
y2=[4,8,12,16]
plt.axis([0,10,0,20])
plt.title("multi line plot")
plt.plot(x1,y1,"r-.o",label="1st line")
plt.plot(x2,y2,"g--s",label="2nd line")
plt.xlabel("horizontal axis")
plt.ylabel("vertical axis")
plt.legend(loc="best")  #it is necessary for label
plt.show()
```

```python
plt.title("multi line plot")
plt.plot([1,2,3,4],[2,4,6,8],"r-.o",label="1st line")
plt.plot([3,6,9,12],[4,8,12,16],"g--s",label="2nd line")
plt.legend(loc="right")  #it is necessary for label
```
