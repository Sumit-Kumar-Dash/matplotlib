```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=0.05*x*x
plt.ylim(0,5)
plt.plot(x,y)
plt.fill_between(x,y,color="blue",alpha=0.3)  # alpha represent the darkness of plot and it varies from 0 to 2
plt.show()
```

```python
a=[1,2,3,4]
b=[2,4,6,8]
plt.plot(a,b,"b-")
plt.axis([0,6,0,12])
plt.title("plot 1")
plt.fill_between(a,b,color="blue",alpha=1)
plt.show()
```

```python
x=np.arange(0,10)
y1=0.05*x*x
y2=0.09*x*x
plt.ylim(0,5)
plt.plot(x,y1,"ro")
plt.plot(x,y2,"g-")
plt.fill_between(x,y1,y2,color="blue",alpha=0.5)  
plt.show()
```
